````python
def normalize_vec(x):
    """
    Normalitza el vector d'entrada `x` segons la definició
    anterior.
    
    *NO* es poden fer servir bucles ni list-comprehensions
    
    :param x: np.array en forma de vector (unidimensional)
    :return: np.array en forma de vector (unidimensional)
    """
    
    norm_vec = map(lambda y : (y-np.mean(x))/np.std(x), x)
    return(np.array(list(norm_vec)))
    

normalize_vec(np.array([0.04396131, 0.3418877 , 0.75050176, 0.74445241, 0.46609646,
       0.44790707, 0.66703217, 0.54339577]))
````

##  Exercici 2A (Recomanador)

Per fer la recomanació col·laborativa, s'utilitza una funció que ens indiqui quants cops compraria un usuari un producte donat.

Suposarem que fem servir un recomanador basat en usuaris i que la puntuació per a un usuari $u$ i ítem $j$ ve donada per la fòrmula:

$$pred(u, i) = \hat{r}_{u,i} = \frac{\sum_{p\neq u,r_{p,i}>0} sim(u, p)\cdot r_{p,i}}{\sum_{p\neq u,r_{p,i}>0} sim(u, p)}$$

On $r_{u,i}$ indica el nombre de vegades que l'usuari $u$ ha comprat l'l'ítem $i$.

Tens el codi base a continuació

**a) Implementeu una predicció escalada al domini de l'usuari.**

Els usuaris compren en diferent mesura els productes, uns més quantitat, altres menys. Fent servir la següent fórmula escalem la predicció a la mitja i desviació de l'usuari:
$$pred(u, i) = \hat{r}_{u,i} = \sigma_{r_u} \cdot \left[\mu_{r_u} + \frac{\sum_{p\neq u,r_{p,i}>0} sim(u, p)\cdot \frac{r_{p,i}-\mu_{r_p}}{\sigma_{r_p}}}{\sum_{p\neq u,r_{p,i}>0} sim(u, p)}\right]$$
on:

* $\mu_{r_u}$ és la mitjana de compres de l'usuari *u*.
* $\sigma_{r_u}$ és la desviació estàndard de l'usuari *u*.

````python
def prediction(user, item, df, similarities):
    """
    Extreu les similituds i el nombre de vegades que un usuari ha comprat un ítem
    (resp. nombre de vegades que cada usuari ha comprat l'ítem) i crida a la funció 
    anterior per calcular les prediccions.
    
    :param user: ID de l'usuari per la predicció
    :param item: ID de l'ítem per la predicció
    :param df: Dataframe que conté el nombre de vegades que un usuari 
        ha comprat en un `aisle_id`
    :param similarities: Matriu numpy de similituds
    :return : Retorna un escalar (float) amb la predicció
    
    """
    ####################################
    # Codi ja fet, no cal modificar
    
    index = df.index.get_loc(user)
    sims = similarities[index]
    counts = df.loc[:, item]
    
    #
    ####################################
    
    # Modificar o borrar per adaptar a la fórmula 
    return (sims @ counts) / np.sum(sims[counts > 0])
````

````python
df_counts = pd.DataFrame({
    1: [1, 2, 0, 1, 0, 0],
    2: [1, 1, 0, 0, 2, 0],
    3: [1, 2, 3, 1, 0, 1],
    4: [10, 20, 30, 10, 0, 0],
    5: [0, 0, 0, 1, 0, 0],
    6: [1, 1, 1, 1, 1, 2]
}).T

# Predicció amb similitud de Pearson, sense negatius
prediction(1, 2, df_counts, np.maximum(0, df_counts.corr().values))
````

## Exercici 2B (Naive Bayes)

Al lliurament de Naive Bayes vareu implementar la funció *count_words*.
Aqyesta s'encarregava de crear un diccionari on enmagatzenava el numero de cops que cada paraula apareixia en els tweets.

**a) Implementeu la funció *count_bigrams*. Aquesta funció ha de construir un diccionari que contingui tots els bigrames (seqüències de dos tokens) que es puguin trobar al text indicant el nombre d'aparicions.**

```python
{
    bigrama_1 : 88,
    bigrama_2 : 123,
    ...
}
```

````python
####################################
# Codi ja fet, no cal modificar

def memo(f):
    class memodict(dict):
        def __init__(self, f):
            self.f = f
        def __call__(self, *args):
            return self[args]
        def __missing__(self, key):
            ret = self[key] = self.f(*key)
            return ret
    return memodict(f)

@memo    
def normalize(word):
    """
    Funció que donada una paraula la normalitzi
    Exemple: Taller DELS noUS USOS ---> tallers dels nous usos
    
    :param word: paraula a normalitzar
    :return : paraula normalitzada
    """
    
    # ASCII and chars/nums only
    word = ''.join(c for c in word if ord(c) < 128 and (c.isdigit() or c.isalpha()))

    # To lowercase
    word = str(word.strip().lower())
    
    return word

#
####################################
````

````python
from collections import Counter
def count_bigrams(text):
    """
    Funció que ha de construir un diccionari que contingui tots els 
    bigrames que s'han trobat indicant el total de cops que han aparegut.
    
    Les paraules del bigrama han d'estar normalitzades amb la funció anterior
    
    :param text: Text 
    :return : Diccionari amb el format {bigram : n_repeticions, ...}
    """
    
    norm_words = []
    for w in text.split():
        norm_word = normalize(w)
        if norm_word != "":
            norm_words.append(norm_word)
    
    bigrames = []
    for i in range(len(norm_words)-1):
        bigrames.append((norm_words[i],norm_words[i+1]))
    
    count_bigrams = dict(Counter(bigrames))
    return count_bigrams
````

**b) Calculeu la probabilitat de cada bigrama, retorna ordenat de més a menys probable**

La probabilitat es defineix com el nombre de vegadesque apareix un bigrama dividit pel total de bigrames.

Ha de retornar un diccionari de bigrama a probilitat ordenat descendentment per probabilitat

```python
{
    bigrama_1 : 0.x,
    bigrama_2 : 0.y,
    ...
}
```

````python
def bigrams_probabilities(bigrams):
    # Total de bigrames que tenim
    total_bigrams = sum(bigrams.values())
    for bigram, times in bigrams.items():
        bigrams[bigram] = float(times/total_bigrams)
        
    sorted_bigrams = sorted(bigrams.items(), key=lambda x : x[1], reverse=True)
    return sorted_bigrams

bigrams_probabilities(bigrams)
````
